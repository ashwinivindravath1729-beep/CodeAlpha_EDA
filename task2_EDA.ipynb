```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np

print("✅ Libraries imported!")
```

```python
# Step 2: Load the CSV data
import os

# CSV is in Desktop > codealpha project folder
home = os.path.expanduser("~")
csv_path = os.path.join(home, "Desktop", "codealpha project", "books_scraped_data.csv")

df = pd.read_csv(csv_path)

print(f"✅ Data Loaded!")
print(f"Total Rows    : {df.shape[0]}")
print(f"Total Columns : {df.shape[1]}")
print(f"\nColumn Names : {list(df.columns)}")
```

```python
# Step 3: First look at the data
df.head(10)
```

```python
# Step 4: Check data info and missing values
print("📋 Data Info:")
print(df.info())
print("\n🔍 Missing Values:")
print(df.isnull().sum())
```

```python
# Step 5: Price and Rating Statistics
print("📊 Price & Rating Summary:")
print(df[['Price (£)', 'Rating (out of 5)']].describe())
```

```python
# Step 6: Rating Distribution
print("⭐ Rating Distribution:")
print(df['Rating (out of 5)'].value_counts().sort_index())

print("\n📦 Availability Count:")
print(df['Availability'].value_counts())
```

```python
# Step 7: Price category analysis
print("💰 Avg Price per Rating:")
print(df.groupby('Rating (out of 5)')['Price (£)'].mean().round(2))

print("\n🏷️ Price Categories:")
df['Price Category'] = pd.cut(df['Price (£)'], 
                               bins=[0, 20, 40, 60], 
                               labels=['Cheap', 'Medium', 'Expensive'])
print(df['Price Category'].value_counts())
```

```python
# Step 8: EDA Summary
print("=" * 40)
print("   EDA SUMMARY — Books Dataset")
print("=" * 40)
print(f"Total Books Analyzed  : 100")
print(f"Average Price         : £{df['Price (£)'].mean():.2f}")
print(f"Average Rating        : {df['Rating (out of 5)'].mean():.1f} / 5")
print(f"Cheapest Book         : £{df['Price (£)'].min():.2f}")
print(f"Most Expensive Book   : £{df['Price (£)'].max():.2f}")
print(f"All Books In Stock    : Yes ✅")
print(f"Missing Values        : None ✅")
print("=" * 40)
print("✅ Task 2 - EDA Complete!")
```
